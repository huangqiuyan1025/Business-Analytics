```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('shared/Project-3_NYC_311_Calls.pkl')
df = df.set_index(pd.DatetimeIndex(df['Created Date']))
del df['Created Date']
df
```

```python
#What is the average number of daily complaints received in 2022?

df_2022 = df[df.index.year == 2022]
df_2022
```

```python
daily_complaints = df_2022.resample('D')['Unique Key'].count()
avg_complaints_2022 = daily_complaints.mean()
print('Average number of daily complaints received in 2022:', avg_complaints_2022)
```

```python
#On which single date were the maximum number of calls received?
    
calls_by_date = df.groupby(df.index.date).size()
calls_by_date
```

```python
maxday = pd.to_datetime(calls_by_date.idxmax())
maxday
```

```python
#On the date the maximum number of calls were received, what was the most important complaint type?

max_day_calls = df[df.index.date == maxday.date()]
max_day_calls
```

```python
max_day_calls['Complaint Type'].value_counts().idxmax()
```

```python
#Group the data by months, and identify the month that historically has the fewest number of calls.

calls_by_month = df.groupby(df.index.month).size()
calls_by_month
```

```python
calls_by_month.idxmin()
```

```python
#Resample your time series to a daily frequency.  Perform ETS decomposition based on an additive model.  What is the value of the seasonal component on 2020-12-25

df2 = df.iloc[:,:1]
df2
```

```python
daily_calls = df2.resample('D')['Unique Key'].count()
daily_calls = daily_calls.reset_index(name='Calls')
daily_calls = daily_calls.set_index('Created Date')
daily_calls
```

```python
!pip install statsmodels

import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(daily_calls['Calls'], model = 'additive') 
plt.rcParams['figure.figsize'] = (20, 9)
result.plot()
```

```python
ets = pd.DataFrame({'Calls': daily_calls['Calls'], 
              'trend': result.trend, 
              'seasonality': result.seasonal, 
              'error': result.resid})
ets.loc['2020-12-25', 'seasonality']
```

```python
#Calculate the autocorrelation of the number of daily calls with the number of calls the day prior, ie lag of 1. 

daily_calls['Calls'].autocorr(lag = 1)
```

```python
#Forecast the daily series with a test set of 90 days using the Prophet library.  What is your RMSE on your test set?

!pip install prophet
from prophet import Prophet
```

```python
daily_calls = pd.DataFrame(daily_calls)
daily_calls = daily_calls.reset_index().rename(columns = {'Created Date': 'ds', 'Calls': 'y'})
daily_calls
```

```python
y_train = daily_calls[:-90]
y_test = daily_calls[-90:]
y_train,y_test
```

```python
model = Prophet()
model.fit(y_train)
future = model.make_future_dataframe(periods = 90, freq = 'd')
future.tail()
```

```python
!pip install scikit-learn
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

forecast = model.predict(future)
actual = y_test['y'].values

predicted = forecast[-90:]['yhat'].values
rmse = sqrt(mean_squared_error(actual, predicted))
rmse
```
